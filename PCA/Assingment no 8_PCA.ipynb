```python
#Import libraries
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
import scipy.cluster.hierarchy as sch
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score as sil
import warnings
warnings.filterwarnings("ignore")
```

# Scaling of Data

```python
sc=MinMaxScaler()
sc.fit(df_array)
new_df_array=sc.transform(df_array)
```

```python
new_df_array
```

# PCA implementation

```python
pca=PCA()
PCA_df=pca.fit_transform(new_df_array)
```

```python
PCA_df
```

```python
pca.components_
```

```python
Var=pca.explained_variance_ratio_
```

```python
Var
```

```python
plt.bar(range(1,len(Var)+1),Var)
```

```python
#cumulative variance
Var1=np.cumsum(np.round(Var,decimals=4)*100)
Var1
```

```python
# Variance plot for PCA components obtained 
plt.plot(Var1,color="red")
```

```python
new_df.shape
```

```python
PCA_df.shape
```

```python
wineFinal=pd.concat([pd.DataFrame(PCA_df[:,0:3],columns=['pc1','pc2','pc3']),df[["Type"]]],axis=1)
```

```python
wineFinal
```

# Herarchical Clustering

```python
# Create dendrogram of Normalized data for all methots 
for methods in ['single','complete','average']: 
     plt.figure(figsize =(20, 6))
     plt.title('Visualising the data, Method- {}'.format(methods)) 
     Dendrogram = sch.dendrogram(sch.linkage(new_df_array, method = methods,optimal_ordering=False))
```

```python
nc=[2,3,4,5,6,7,8,9]
sc_score=[]    
for i in nc:
    hc = AgglomerativeClustering(n_clusters=i, affinity = 'euclidean', linkage = 'complete')

    y=hc.fit_predict(new_df_array)
    sc_score.append(sil(new_df_array,y))
```

```python
plt.plot(nc,sc_score,color="black")
plt.scatter(nc,sc_score,color="red")
plt.show()
```

```python
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
hclusters
y_hc=hclusters.fit_predict(new_df_array)
```

# Kmeans Clustering

```python
nc=[2,3,4,5,6,7,8,9]
iner=[]    
for i in nc:
    km=KMeans(i)
    km.fit(new_df_array)
    iner.append(km.inertia_)
```

```python
plt.plot(nc,iner,color="black")
plt.scatter(nc,iner,color="red")
plt.show()
```

```python
model_kmeans = KMeans(n_clusters=3) 
model_kmeans.fit(new_df_array)
```

```python
wdf3=df
```

```python
wdf3["Norm_KMclustering"]=pd.Series(model_kmeans.labels_)
```
